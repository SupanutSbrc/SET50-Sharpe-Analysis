The Set Up

```python
#Close Price Year End
tv = TvDatafeed()
df = tv.get_hist(symbol='SET:SET50', exchange='SET', interval=Interval.in_monthly, n_bars=60)
set50 = df[['close']]
set50.rename(columns = {'close':'set50'}, inplace=True)
for i in range (len(name)):
    df = tv.get_hist(symbol=name[i], exchange='SET', interval=Interval.in_monthly, n_bars=60)
    df = df[['close']]
    set50 = pd.merge(set50, df, left_index=True, right_index=True, how='outer')
    set50.rename(columns = {'close':name[i]}, inplace=True)
close = set50[set50.index.month == 12]
last_row = set50.iloc[[-1]]
close = pd.concat([close, last_row])
years = ['2019','2020', '2021', '2022', '2023','2024']
close = close.reset_index(drop=True)
close.index = years
close = close[name]
close = close.tail(4)
close
```

```python
#Open Price Year
df = tv.get_hist(symbol='SET:SET50', exchange='SET', interval=Interval.in_monthly, n_bars=60)
set50 = df[['close']]
set50.rename(columns = {'close':'set50'}, inplace=True)
for i in range (len(name)):
    df = tv.get_hist(symbol=name[i], exchange='SET', interval=Interval.in_monthly, n_bars=60)
    df = df[['open']]
    set50 = pd.merge(set50, df, left_index=True, right_index=True, how='outer')
    set50.rename(columns = {'open':name[i]}, inplace=True)
open = set50[set50.index.month == 3]
open = open.reset_index(drop = True)
years = ['2020', '2021', '2022', '2023','2024']
open.index = years
open = open[name]
open = open.tail(4)
open
```

```python
#Return (Close - Open of each year)
returns = pd.DataFrame()
returns = close - open
returns
```

The Strategy

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

strategy = pd.DataFrame()
strategy = ((assets) + (revenue/assets)/2)

####Sharpe > 1 and is currently profiting
#assets/liabilities
#(assets/liabilities) + (revenue/assets)/4 - (liabilities/assets)
#((assets/liabilities)/2 + (revenue/assets)/4) / (liabilities/assets)
# 1 / (liabilities/assets)

####Sharpe > 1 but is currently losing
#((assets) + (revenue/assets)/2)

####Shart < 1 but is currently profiting
#net_profit/mcap + equity/assets (sharpe 0.84)
#((assets/liabilities) + (revenue/assets)/2)
ninvest = 5
notional = 1000000
strategy
```

```python
#close 2020, backtest return of 2021, 2022, 2023, and return in 2024 so far
years = ['2021','2022','2023']
#strategy.drop(columns=['SET:INTUCH'], inplace=True)
performance = pd.DataFrame({'year':years})
yearly_return = []
for years in years:
    temp = strategy.loc[years].nlargest(ninvest).index
    top_df = strategy.loc[[years], temp]
    top_df =  top_df.T.reset_index()
    total = np.sum(top_df[years])
    top_df['weight'] = top_df[years] / total
    top_df['allocation'] = notional * top_df['weight']

    priz = []
    for i in range(len(top_df)):
        price = open[top_df['index'][i]][years]
        priz.append(price)

    top_df['price'] = np.array(priz)
    top_df['n.Stock'] = np.floor(top_df['allocation'] / top_df['price'])

    ret = []
    for i in range(len(top_df)):
        retur = returns[top_df['index'][i]][years]
        ret.append(retur)

    top_df['return'] = np.array(ret)
    top_df['totalReturn'] = top_df['return'] * top_df['n.Stock']
    perf = np.sum(top_df['totalReturn'])
    yearly_return.append(perf)

    print(top_df)
performance['Yearly_Return'] = np.array(yearly_return)
performance['% Return'] = round(performance['Yearly_Return']/notional * 100,2)
print(performance)
mean = np.mean(performance['% Return'])
mean
std = np.std(performance['% Return'])
sharpie = (mean - rrf) / std
total_return = np.sum(performance['Yearly_Return'])
total_pct_return = total_return/notional
summary = pd.DataFrame({
    'Sharpe': [sharpie],
    'Total Return': [total_return],
    'Total Percentage Return': [total_pct_return]
})
summary = summary.round(2)
summary
```
